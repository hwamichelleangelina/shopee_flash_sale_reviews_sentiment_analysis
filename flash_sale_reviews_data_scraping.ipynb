```python
## Data Scraping
def scrape_shopee_reviews(item_id, shop_id, max_pages=300):
    reviews = []
    base_url = "https://shopee.co.id/api/v2/item/get_ratings"

    for page in range(1, max_pages + 1):
        params = {
            "itemid": item_id,
            "shopid": shop_id,
            "limit": 50,
            "offset": (page - 1) * 50,
        }

        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        }

        response = requests.get(base_url, params=params, headers=headers)
        if response.status_code == 200:
            data = response.json()
            if "data" in data and "ratings" in data["data"]:
                reviews_list = data["data"]["ratings"]
                #print(f"Page {page} - Total reviews fetched: {len(reviews_list)}")  # Debugging jumlah ulasan
                for review in reviews_list:
                    reviews.append({
                        "username": review.get("author_username", ""),
                        "rating": review.get("rating_star", ""),
                        "comment": review.get("comment", ""),
                        "time": review.get("ctime", ""),
                    })
            else:
                print(f"Page {page} - No reviews found.")
                break
        else:
            print(f"Failed to fetch data for page {page}, status code: {response.status_code}")
            break

        time.sleep(2)

    return reviews

# Identitas Toko Shopee
shop_id = 346000752
item_id = 13217782061

reviews_data = scrape_shopee_reviews(item_id, shop_id, max_pages=65)
print(f"Jumlah komentar yang berhasil diambil: {len(reviews_data)}")

## Simpan data ke CSV
df = pd.DataFrame(reviews_data)
df["time"] = pd.to_datetime(df["time"], unit="s")
df.to_csv("miband_shopee_reviews.csv", index=False)
print("Data disimpan di miband_shopee_reviews.csv.")
```
